```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import glob

plt.rcParams['figure.figsize'] = (10.0, 8.0)
```

```python
data = pd.read_csv("runs/mc-InvertedPendulum-v1-0.01-0.001-True-exp_id=NA-run_id=NA-learnStd=False-seed=7316-use_target=False-1557724366-num_samples=100")
```

```python
def get_df_for_metric(regex, metric_name):
    files = glob.glob(f"runs/{regex}")
    print(f"Found {len(files)} files for regex.")
    dfs = [pd.read_csv(file) for file in files]
    dfs = list(map(lambda x: x.loc[x['metric_name'] == metric_name], dfs))
    df = dfs[0]; i=1
    for df2 in dfs[1:]:
        df2.columns = [f"{column}_{i}" for column in df2.columns]
        df = df.merge(df2, how="outer", on="episode")
        i+=1
    return df
```

```python
get_df_for_metric("*reinforce*Pendulum*", "total_reward").columns
```
